```python
df = pd.read_csv("./Roaming_Data_CSV.csv")
print(df.shape)
df
```

### Filters: Mobile originated Call, Mobile Terminating Call, Short Message Originating Call, GPRS

```python
MOC = df.loc[(df['CALL TYPE DESCRIPTION'].str.match('Mobile originated Call'))]
MOC
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.figure(figsize=(25,10))
plt.bar(MOC['TRANSDATE'], MOC['VOLUME'],width=0.4)
plt.xticks(rotation = 90)
plt.title("Mobile Originated Call")
plt.xlabel("Date")
plt.ylabel("Time (Second)")
plt.show()
```

```python
MTC = df.loc[(df['CALL TYPE DESCRIPTION'].str.match('Mobile Terminating Call'))]
MTC
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.figure(figsize=(25,10))
plt.bar(MTC['TRANSDATE'], MTC['VOLUME'],width=0.4)
plt.xticks(rotation = 90)
plt.title("Mobile Terminating Call")
plt.xlabel("Date")
plt.ylabel("Time (Second)")
plt.show()
```

```python
SMOC = df.loc[(df['CALL TYPE DESCRIPTION'].str.match('Short Message Originating Call'))]
SMOC
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.figure(figsize=(25,10))
plt.bar(SMOC['TRANSDATE'], SMOC['VOLUME'],width=0.4)
plt.xticks(rotation = 90)
plt.title("Short Message Originating Call")
plt.xlabel("Date")
plt.ylabel("Count")
plt.show()
```

```python
GPRS = df.loc[(df['CALL TYPE DESCRIPTION'].str.match('GPRS'))]
GPRSGB= GPRS['VOLUME']/1073741824
GPRSGB
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
plt.figure(figsize=(25,10))
plt.bar(GPRS['TRANSDATE'],GPRSGB,width=0.4)
plt.xticks(rotation = 90) 
plt.title("General Packet Radio Service (GPRS)")
plt.xlabel("Date")
plt.ylabel("Giga Bytes")
plt.show()
```
