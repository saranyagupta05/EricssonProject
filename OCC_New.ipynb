### Separating initial, termination and update requests 

```python
copy_in = df_occ12.loc[df_occ12[1].str.match('occ1') & df_occ12[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr = df_occ12.loc[df_occ12[1].str.match('occ1') & df_occ12[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud = df_occ12.loc[df_occ12[1].str.match('occ1') & df_occ12[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]

copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df12['occ1_initial']=copy_in[6]
df12['occ1_termination']=copy_tr[6]
df12['occ1_update']=copy_ud[6]

copy_in= df_occ12.loc[df_occ12[1].str.match('occ2')&df_occ12[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr= df_occ12.loc[df_occ12[1].str.match('occ2')&df_occ12[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud= df_occ12.loc[df_occ12[1].str.match('occ2')&df_occ12[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]
copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df12['occ2_initial']=copy_in[6]
df12['occ2_termination']=copy_tr[6]
df12['occ2_update']=copy_ud[6]
df12
```

```python
copy_in = df_occ34.loc[df_occ34[1].str.match('occ3') & df_occ34[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr = df_occ34.loc[df_occ34[1].str.match('occ3') & df_occ34[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud = df_occ34.loc[df_occ34[1].str.match('occ3') & df_occ34[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]

copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df34['occ3_initial']=copy_in[6]
df34['occ3_termination']=copy_tr[6]
df34['occ3_update']=copy_ud[6]

copy_in= df_occ34.loc[df_occ34[1].str.match('occ4')&df_occ34[5].str.match('Diameter Front-End.Gy-CCR-Initial.request')]
copy_tr= df_occ34.loc[df_occ34[1].str.match('occ4')&df_occ34[5].str.match('Diameter Front-End.Gy-CCR-Termination.request')]
copy_ud= df_occ34.loc[df_occ34[1].str.match('occ4')&df_occ34[5].str.match('Diameter Front-End.Gy-CCR-Update.request')]
copy_in.set_index(0,inplace=True)
copy_tr.set_index(0,inplace=True)
copy_ud.set_index(0,inplace=True)
df34['occ4_initial']=copy_in[6]
df34['occ4_termination']=copy_tr[6]
df34['occ4_update']=copy_ud[6]
df34
```

```python
df34['Traffic']=df34['occ3_initial']+df34['occ3_termination']+df34['occ3_update']+df34['occ4_initial']+df34['occ4_termination']+df34['occ4_update']
df34['Traffic'][0]=0 # all values are NaN 
df34.reset_index(drop=True, inplace=True)
uniqueValues =df_occ34[0].unique()
df34['Date']= uniqueValues
df34[['Date','Time']] = df34.Date.str.split(expand=True)
df34
```

```python
uniqueValues =df34['Date'].unique()
data = pd.DataFrame(uniqueValues)

def traffic(a):
    return (df12.loc[df12['Date']==a]['Traffic'].sum()+df34.loc[df34['Date']==a]['Traffic'].sum())

data['Count']=data.apply(lambda x: traffic(x[0]),axis=1)
data
```

```python
# Plotting 
import matplotlib.pyplot as plt 
plt.figure(figsize=(25,10))
plt.bar(data[0], data['Count'],width=0.5)
plt.xlabel("Date")
plt.xticks(rotation = 90)
plt.title("OCC Traffic")
plt.show()
```
